```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import math
```

```python
data = pd.read_csv("data/sample_data_features.csv")
features = list(data.columns[4:])
```

```python
fig, axes = plt.subplots(ncols=2, nrows=math.ceil(len(features)/2), figsize=(22, len(features)*1.4))
plt.subplots_adjust(hspace=0.35)
plt.rcParams.update({"font.size": 14})
for ax, aspect in zip(axes.reshape(-1), features):
    sns.boxplot(data=data, x=aspect, y="source", linewidth=2.5, zorder=5, showfliers=False, ax=ax)
    sns.stripplot(data=data, x=aspect, y="source", hue="source", edgecolor='gray', linewidth=1, zorder=10, ax=ax)
    ax.set_axisbelow(True) 
    ax.grid(alpha=0.5)
    ax.get_legend().remove()
```
